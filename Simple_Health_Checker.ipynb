```python
KNOWLEDGE_BASE = {
    "Common Cold": {"runny nose", "sore throat", "sneezing", "headache"},
    "Influenza (Flu)": {"fever", "body aches", "fatigue", "headache", "sore throat"},
    "Strep Throat": {"sore throat", "fever", "difficulty swallowing", "red spots on throat"},
    "Allergies": {"sneezing", "runny nose", "itchy eyes", "post-nasal drip"},
    "Migraine": {"severe headache", "nausea", "sensitivity to light"},
    "Gastroenteritis (Stomach Flu)": {"nausea", "vomiting", "diarrhea", "stomach pain"}
}
def get_user_symptoms():
    print("Please enter your symptoms, separated by commas (e.g., fever, runny nose, headache).")
    user_input = input("Enter your symptoms: ").strip().lower()
    symptoms_list = [symptom.strip() for symptom in user_input.split(',') if symptom.strip()]
    return set(symptoms_list)

def check_symptoms(user_symptoms):
    if not user_symptoms:
        return []
    condition_scores = {}
    for condition, required_symptoms in KNOWLEDGE_BASE.items():
        matched_symptoms = user_symptoms.intersection(required_symptoms)
        match_count = len(matched_symptoms)
        if match_count > 0:
            condition_scores[condition] = match_count
    sorted_conditions = sorted(
        condition_scores.items(),
        key=lambda item: item[1],
        reverse=True
    )
    return sorted_conditions

def display_results(results, user_symptoms):
    print("Potential Conditions")
    strong_results = [item for item in results if item[1] >= 2] 
    if strong_results:
        print("\nBased on your symptoms, here are the top possibilities:")
        for i, (condition, score) in enumerate(strong_results[:3]):
            print(f"#{i+1}: **{condition}** (Matched {score} key symptoms)")
            required_symptoms = KNOWLEDGE_BASE[condition]
            matched = user_symptoms.intersection(required_symptoms)
            print(f"   -> Symptoms matched: {', '.join(matched).capitalize()}")
    else:
        print("\n Your symptoms do not strongly match a specific condition in my simple list.")
        if user_symptoms:
             print(f"You entered: {', '.join(user_symptoms).capitalize()}")

    print("\n--- Disclaimer ---")
    print("This is a simple program and NOT a medical diagnosis. Always consult a doctor for health concerns.")

if __name__ == "__main__":
    symptoms = get_user_symptoms()
    results = check_symptoms(symptoms)
    display_results(results, symptoms)
```
